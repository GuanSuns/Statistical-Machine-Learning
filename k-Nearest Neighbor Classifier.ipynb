# k-Nearest Neighbor

- Most codes are based on this kNN tutorial: https://scikit-learn.org/stable/auto_examples/neighbors/plot_classification.html

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors
from scipy.spatial import Voronoi, voronoi_plot_2d

X = np.array([])
Y = np.array([])
```

## Dataset

```python
def use_data_A():
    global X
    global Y
    
    X = np.array([[3, 4], [5, 6], [7, 4]])
    X = X.astype('float64') 
    Y = np.array([1, 1, -1])
    
    
def use_data_B():
    global X
    global Y
    
    X = np.array([[5, 7], [7, 5], [5, 3], [7, 1]])
    X = X.astype('float64') 
    Y = np.array([1, 1, -1, -1])
    
    
def use_data_C():
    global X
    global Y
    
    X = np.array([[3, 9], [3, 6], [3, 4], [5, 8], [3, 2], [3, 0], [7, 4], [7, 6]])
    X = X.astype('float64') 
    Y = np.array([1, 1, 1, 1, -1, -1, -1, -1])

def use_data_D():
    global X
    global Y
    
    X = np.array([[10, 10], [8, 8], [7, 5], [5, 7], [3, 1], [5, 0], [5, 3], [7, 1]])
    X = X.astype('float64') 
    Y = np.array([1, 1, 1, 1, -1, -1, -1, -1])
```

```python
use_data_A()
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='autumn')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(np.arange(0, 11, 1.0))
plt.yticks(np.arange(0, 11, 1.0))
plt.grid()
```

```python
use_data_B()
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='autumn')
plt.xticks(np.arange(0, 11, 1.0))
plt.yticks(np.arange(0, 11, 1.0))
plt.grid()
```

## Visualize the Boundary

```python
def visualize_boundary(X, Y, n_neighbors=1, weights='uniform', step_size_mesh = 0.02):
    """
    Codes are modified based on the sklearn documentation
    """
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    model = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    model.fit(X, Y)
    
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = 0, 10.0
    y_min, y_max = 0, 10.0
    xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size_mesh),
                         np.arange(y_min, y_max, step_size_mesh))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=cmap_bold,
                edgecolor='k', s=80)
    
    # If n_neighbors is 1, also show the Voronoi cells
    ax = plt.gca()
    if n_neighbors == 1:
        vor = Voronoi(X)
        voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='black', line_width=2, line_alpha=0.6, point_size=2)
    
    ax.set_xlim((0, 10))
    ax.set_ylim((0, 10))
    plt.xticks(np.arange(0, 11, 1.0))
    plt.yticks(np.arange(0, 11, 1.0))
    plt.grid()
```

```python
# Problem A
use_data_A()

visualize_boundary(X, Y)
plt.title("Figure (a). k = 1, weights = 'uniform'")
```

```python
# Problem B
use_data_B()

visualize_boundary(X, Y)
plt.title("Figure (b). k = 1, weights = 'uniform'")
```

```python
# Problem C
use_data_C()

visualize_boundary(X, Y)
plt.title("Figure (c). k = 1, weights = 'uniform'")
```

```python
# Problem D
use_data_D()

visualize_boundary(X, Y)
plt.title("Figure (d). k = 1, weights = 'uniform'")
```
